```python
def scrape_stupid_site():
    base = 'https://www.whiskystats.net/whiskydetail/?whisky_id='
    num_bottles = 59736
    bottles = []
    for i in tqdm.tqdm_notebook(range(1, num_bottles)):
        temp = []
        browser.visit(base + str(i))
        sleep(.1)
        soup = BeautifulSoup(browser.html, 'html.parser')
        temp = strip_list(soup.find_all('td', 'st-val'))
        temp.append(soup.find_all('td', 'value')[-1].text)
        bottles.append(temp)
#         print(temp)
    return pd.DataFrame(bottles)
```

```python
base = 'https://www.whiskystats.net/whiskydetail/?whisky_id='
num_bottles = 59736
bottles = []
with open('whisky-stats-shit4.csv', 'a') as f:
    wr = csv.writer(f, dialect='excel')
    for i in tqdm.tqdm_notebook(range(3000, num_bottles)):
        temp = []
        browser.visit(base + str(i))
        sleep(.5)
        try:
            soup = BeautifulSoup(browser.html, 'html.parser')
            temp = strip_list(soup.find_all('td', 'st-val'))
            temp.append(soup.find_all('td', 'value')[-1].text)
            temp.append(soup.find('h1', 'entry-title').text.strip())
            temp.append(soup.find('h2', {'id':'whisky-subtitle'}).text.strip())
            bottles.append(temp)
    #         print(temp)
        except Exception as e:
            print(e)
        if(i%50 == 0):
            print('saving...')
            wr.writerows(bottles)
            bottles = []
```

```python
def strip_list(l):
    result = []
    for i in l:
        result.append(i.text)
    return result
        
```
