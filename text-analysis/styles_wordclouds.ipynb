```python
def replace_text(item):
    item.replace("  ", " ").replace(".", "").replace(",", "").replace("!", "").replace("'", "").replace('"', "").replace(";", "").replace(":", "").replace("(", "").replace(")", "").replace("/", "").replace("fruits", "fruit")
    
    
```

```python
text=sms_df["descriptions"].str.cat(sep=' ').lower().strip()
text=text.replace("  ", " ").replace(".", "").replace(",", "").replace("!", "").replace("'", "").replace('"', "").replace(";", "").replace(":", "").replace("(", "").replace(")", "").replace("/", "").replace("fruits", "fruit")
clean_text=text.split(" ")    
for word in clean_text:
    if word in sr:        
        clean_text.remove(word)
tokens=[]
for word in clean_text:
    tokens.append(word)
clean_tokens = tokens[:]
for token in tokens:
    if token in sr:
        clean_tokens.remove(token)
sms_freq = nltk.FreqDist(clean_tokens)
sms_freq

```
