```python
from splinter import Browser 
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
from tqdm import tqdm_notebook
```

```python
browser = Browser('chrome')
```

```python
browser.visit('https://www.whiskystats.net/whisky-database/#')
```

```python
soup = BeautifulSoup(browser.html, 'html.parser')
```

```python
def load_page():
    num_times = int(int(soup.find('div', {'id':'result-nof'}).text.split()[0])/100)
    btn = browser.find_by_text('Load More').first
    i = num_times
    while i > 0:
        print(num_times - i)
        try:
            btn.click()
        except Exception as e:
            print(e)
            sleep(2)
#             i += 1
        sleep(.2)
        i -= 1
```

```python
def load_page_2():
    num_times = int(int(soup.find('div', {'id':'result-nof'}).text.split()[0])/100)
    btn = browser.find_by_text('Load More').first
    for i in tqdm_notebook(range(num_times)):
        btn.click()
        sleep(5)
```

```python
load_page_2()
```
