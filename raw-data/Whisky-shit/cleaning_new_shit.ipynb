```python
import pandas as pd
```

```python
df1 = pd.read_csv('new-shit-Campbeltown.csv')
```

```python
df2 = pd.read_csv('new-shit-Highlands.csv')
```

```python
df3 = pd.read_csv('new-shit-Islands.csv')
```

```python
df4 = pd.read_csv('new-shit-Islay.csv')
```

```python
df5 = pd.read_csv('new-shit-Japan.csv')
```

```python
df6 = pd.read_csv('new-shit-Lowlands.csv')
```

```python
df7 = pd.read_csv('new-shit-Speyside.csv')
```

```python
df8 = pd.read_csv('new-shit-us.csv')
```

```python
df_uk = pd.concat([df1, df2, df3, df4, df6, df7])
```

```python
df_us = df8.copy()
```

```python
df_uk.dropna(subset = ['vintage'], inplace = True)
```

```python
df_uk['price'] = df_uk['price'].apply(lambda x: x.split()[0]).astype('int32')
```

```python
df_uk['vol']= df_uk['vol'].apply(lambda x: x.replace('%', '')).astype('float32')
```

```python
df_uk['vintage'] = df_uk['vintage'].astype('int32')
```

```python
df_uk.to_csv('scottish_whisky_stats.csv', index = False)
```
