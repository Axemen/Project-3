```python
import pandas as pd
```

```python
df1 = pd.read_csv('whisky-stats-shit1.csv', encoding = 'ISO-8859-1')
df2 = pd.read_csv('whisky-stats-shit2.csv', encoding = 'ISO-8859-1')
df3 = pd.read_csv('whisky-stats-shit3.csv', encoding = 'ISO-8859-1')
df4 = pd.read_csv('whisky-stats-shit4.csv', encoding = 'ISO-8859-1')
```

```python
names = ['age', 'vintage', 'bottled','avp', 'size', 'cask', 'num_bottles', 'price', 'name', 'bottling']
```

```python
df1.set_axis(names, axis=1, inplace = True)
```

```python
df2.head()
```

```python
df2.set_axis(names, axis=1, inplace = True)
```

```python
df3.set_axis(names, axis=1, inplace = True)
```

```python
df4.set_axis(names, axis=1, inplace = True)
```

```python
dfs = [df1, df2, df3, df4]
```

```python
df = pd.concat(dfs)
```

```python
df = df.dropna(subset = ['vintage', 'price']).drop(columns = ['cask', 'num_bottles'])
```

```python
df['price'] = df['price'].apply(lambda x: x.split()[0]).astype('int64')
```

```python
df['size'] = df['size'].apply(lambda x: x.split()[0]).astype('int64')
```

```python
df['price'] = df['price'].apply(lambda x: x * 1.11)
```

```python
df['bottler'] = ''
df['loc/type'] = ''

for i in range(len(df)):
    s = df.iat[i, 7].split(',')
    df.iat[i, 8] = s[0]
    df.iat[i, 9] = s[1]
```
